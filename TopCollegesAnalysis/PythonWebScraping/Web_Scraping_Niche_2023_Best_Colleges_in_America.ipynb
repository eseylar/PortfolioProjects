```python
#@title IMPORT LIBRARIES
from bs4 import BeautifulSoup as bs
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Use the code block below to scrape data straight from Niche's website. (Note: niche.com requires Javascript, so this code must be executed in a working JS envrionment, perhaps by creating a Sellinum envrionment in Python`.)

```python
#@title PARSING FROM NICHE URLS

pg = 1
top_schools = []

while pg != 11: # generate the first ten pages of results

    url = f"https://www.niche.com/colleges/search/best-colleges/?page={pg}"
    response = requests.get(url)
    html = response.content
    soup = bs(html, "lxml")

    for x in soup.find_all("div", class_="search-result"):
        school = {}

    #School ranking (1-250)   
        rank = x.find("div", class_="MuiTypography-root MuiTypography-subtitle2 search-result-badge nss-tcks16")
        rank_str = rank.get_text ('aria-label')
        rank_int = int(rank_str[1:rank_str.index(" ")])
        school["rank"] = rank_int

    #School name   
        name = x.find("h2", class_="MuiTypography-root MuiTypography-h3 MuiLink-root MuiLink-underlineHover nss-ihhgee")
        school["name"] = name.get_text ('aria-label')

    #School location as City, State    
        city_state = x.find("p", class_="MuiTypography-root MuiTypography-body2 nss-1l1273x")
        city_state_str = city_state.get_text ("")
        city_state_str = city_state_str[:-1]
        school["city_state"] = city_state_str

    #School location as city
        city = city_state_str.split(", ")[0]
        school["city"] = city

    #School location as state
        state = city_state_str.split(", ")[1]
        school["state"] = state

    #School acceptance percentage rate, as integer
        acceptance = x.find_all("p", class_="MuiTypography-root MuiTypography-body2 nss-1l1273x")[2]
        acceptance_str = acceptance.get_text()
        acceptance_str = acceptance_str[15:-1]
        acceptance_int = int(acceptance_str)
        school["acceptance"] = acceptance_int

    #School "net price" after average aid, as integer
        price = x.find_all("p", class_="MuiTypography-root MuiTypography-body2 nss-1l1273x")[3]
        price_str = price.get_text()
        price_str = price_str[11:]
        try:
            price_int = int(price_str.replace(',', ''))
        except:
            price_int = None
        school["price"] = price_int

    #School SAT range, as lower_value - higher_value
        SAT_range = x.find_all("p", class_="MuiTypography-root MuiTypography-body2 nss-1l1273x")[4]
        SAT_range_str = SAT_range.get_text()
        SAT_range_str = SAT_range_str[10:]
        if SAT_range_str == "—":
            SAT_range_str = None
        else:
            SAT_range_str = SAT_range_str
        school["SAT_range"] = SAT_range_str

    #School SAT lower limit, as integer
        try:
            SAT_low_int = int(SAT_range_str.split("-")[0])
        except:
            SAT_low_int = None
        school["SAT_low"] = SAT_low_int

    #School SAT upper limit, as integer
        try:
            SAT_high_int = int(SAT_range_str.split("-")[1])
        except:
            SAT_high_int = None
        school["SAT_high"] = SAT_high_int

    #Append top_schools with information from each iteration of the x function (for each school)
        top_schools.append(school)

#Add 1 to page counter to cycle to next page
    pg = pg + 1



#Turn the list of items into a data-frame
schools_df = pd.DataFrame(top_schools)
schools_df
```
